### Importing the required libraries

```python
import numpy as np
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt
%matplotlib inline
```

### Creating x,y1,y2 lines and the plot space

```python
np.random.seed(5)
x = np.linspace(1,40,10)
y1 = x/(np.random.rand(10)*2)
y2 = x/(np.random.rand(10)*3)
line1 = Line2D(x,y1)
line2 = Line2D(x,y2)

fig, ax = plt.subplots()

ax.add_line(line1)
ax.add_line(line2)
```

### Scaling the scale of the graph so that the lineplots are visible

```python
ax.autoscale()
fig
```

```python
ax.lines
```

### Changing the colour of one line for better visualisation

```python
ax.lines[0].set_color('r')
fig
```

### help() gives us all the information about customisation

```python
help(fig)
```

### Changing the size of the plot

```python
fig.set_size_inches(15,5)
fig
```

```python
l1 = ax.lines[0]
l2 = ax.lines[1]
```

### Changing the opacity of line l1

```python
l1.set_alpha(0.3)
fig
```

### 1) Changing width of the lines
### 2) Making 1st line dash

```python
#1
for line in ax.lines:
    line.set_lw(2)
#2
l1.set_linestyle('dashed')
fig
```

### Setting marker for the 2nd line

```python
l2.set_marker('*')
l2.set_markersize(10)
l2.set_markerfacecolor('white')
fig
```

### Setting the labels

```python
ax.set_title('Some line chart')
ax.set_xlabel('This is x axis')
ax.set_ylabel('This is y axis')
fig
```

### Saving the figure as png

```python
fig.savefig('artist1.png', dpi=350, transparent=False)
```
